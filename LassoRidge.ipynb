```python
import pandas as pd

computer_data = pd.read_csv("D:\\360DigiTMG\\DataScience\\27. Lasso-Ridge Regression\\Assignment Q\\Computer_Data (1).csv")
computer_data = computer_data.iloc[:, 1:]
computer_data.head()
```

```python
computer_data_obj = computer_data.iloc[:, 5:8]
```

```python
computer_data_obj
```

```python
computer_data.drop(["cd", "multi", "premium"], axis=1, inplace=True)
```

```python
computer_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
computer_data = pd.DataFrame(mms.fit_transform(computer_data), columns=computer_data.columns)
```

```python
computer_data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbl = LabelEncoder()
```

```python
computer_data_obj.cd = lbl.fit_transform(computer_data_obj.cd)
```

```python
computer_data_obj.multi = lbl.fit_transform(computer_data_obj.multi)
```

```python
computer_data_obj.premium = lbl.fit_transform(computer_data_obj.premium)
```

```python
computer_data_obj
```

```python
computer_data = pd.concat([computer_data, computer_data_obj], axis=1)
```

```python
computer_data
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.23)
lasso.fit(computer_data.iloc[:, 1:], computer_data.price)
lasso.intercept_
```

```python
lasso.coef_
```

```python
pred_lasso = lasso.predict(computer_data.iloc[:, 1:])
pred_lasso
```

```python
import numpy as np
```

```python
rmse1 = np.sqrt(np.mean((pred_lasso - computer_data.price)**2))
rmse1
```

```python
import pandas as pd
```

```python
toyota = pd.read_csv("D:\\360DigiTMG\\DataScience\\27. Lasso-Ridge Regression\\Assignment Q\\ToyotaCorolla (1).csv",  encoding= 'unicode_escape')
```

```python
toyota.head()
```

```python
toyota.drop(["Model"], axis=1, inplace=True)
```

```python
toyota
```

```python
toyota_obj = toyota.select_dtypes(["object"])
```

```python
toyota_obj
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbl = LabelEncoder()
```

```python
toyota_obj.Fuel_Type = pd.Series(lbl.fit_transform(toyota_obj.Fuel_Type))
```

```python
toyota_obj.Color = pd.Series(lbl.fit_transform(toyota_obj.Color))
```

```python
toyota_obj
```

```python
toyota.drop(["Fuel_Type", "Color"], axis=1, inplace=True)
```

```python
toyota = pd.concat([toyota, toyota_obj], axis=1)
```

```python
toyota
```

```python
toyota.drop(["Id"], axis=1, inplace=True)
```

```python
toyota
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5 ,10, 20]}

ridge_reg = GridSearchCV(ridge, params, scoring='r2', cv=5)
ridge_reg.fit(toyota.iloc[:, 1:], toyota.Price)
```

```python
ridge_reg.best_params_
```

```python
ridge_pred = ridge_reg.predict(toyota.iloc[:, 1:])
ridge_pred
```

```python
np.sqrt(np.mean((ridge_pred-toyota.Price)**2))
```
